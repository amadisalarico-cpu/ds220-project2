```python
import pandas as pd
!wget https://github.com/pandas-dev/pandas/raw/main/doc/data/titanic.csv
df = pd.read_csv('titanic.csv')
print("âœ… DATA LOADED!")
df.head()

print("\n" + "="*80)
print("ðŸ”¥ TITANIC SURVIVAL ANALYSIS - DS 220 PROJECT 2")
print("="*80)
print("Analyzes 1912 RMS Titanic disaster (891 passengers)")
print("Reveals survival patterns â†’ Modern risk analysis applications")
print("="*80 + "\n")

# Clean data
df['Age'].fillna(df['Age'].median(), inplace=True)
df.dropna(subset=['Embarked'], inplace=True)
print(f"âœ… CLEANED: {df.shape[0]} rows")
df.info()
```

```python
# Q1: Did women have better survival odds than men?
gender_survive = df.groupby('Sex')['Survived'].mean() * 100
print("Women survived", round(gender_survive['female'],1), "%")
print("Men survived", round(gender_survive['male'],1), "%")
gender_survive.plot(kind='bar', color=['hotpink','lightblue'])
plt.title('Survival Rate: Women vs Men')
plt.ylabel('% Survived')
plt.show()
```

```python
# Q2: Did rich people (1st class) survive more?
class_survive = df.groupby('Pclass')['Survived'].mean() * 100
print("1st Class:", round(class_survive[1],1), "%")
print("2nd Class:", round(class_survive[2],1), "%")
print("3rd Class:", round(class_survive[3],1), "%")
class_survive.plot(kind='bar', color=['gold','silver','brown'])
plt.title('Survival Rate by Ticket Class')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Q3: Were kids more likely to survive?
df['Kid'] = df['Age'] < 18
kids_survive = df.groupby('Kid')['Survived'].mean() * 100
print("Kids (<18):", round(kids_survive[True],1), "% survival")
print("Adults:", round(kids_survive[False],1), "% survival")
kids_survive.plot(kind='bar', color=['orange','gray'])
plt.title('Kids vs Adults Survival')
plt.show()
```

```python
# Q4: Solo travelers vs families?
df['FamilySize'] = df['SibSp'] + df['Parch']
family_survive = df.groupby('FamilySize')['Survived'].mean() * 100
print("Survival by family size:")
print(family_survive.round(1))
family_survive.plot()
plt.title('Survival Rate by Family Size')
plt.ylabel('% Survived')
plt.show()
```

```python
# Q5: Survival by boarding port?
port_survive = df.groupby('Embarked')['Survived'].mean() * 100
print(port_survive.round(1))
port_survive.plot(kind='pie', autopct='%1.1f%%')
plt.title('Survival Rate by Port')
plt.ylabel('')
plt.show()
```
